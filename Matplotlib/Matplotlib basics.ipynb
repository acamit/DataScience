### ploting multiple curves in a diagram

We can just specify both in one cell and we will have both the curves on same graphs

```python
plt.plot(x, np.sin(x), label='sine curve')
plt.plot(x, np.cos(x), label='cosine curve')
plt.legend()
plt.title('Playing with plots')

plt.show()
```

```python
plt.plot(x, np.sin(x), label='sine curve', color='green')
plt.plot(x, np.cos(x), label='cosine curve', color='m') # m is for magenta
plt.legend()
plt.title('Playing with plots')

plt.show()
```

## more plots

```python
random_array= np.random.randn(20) # generate random 20 numbers
```

```python
plt.plot(random_array, color='green') # default line to connect is solid
```

```python
plt.plot(random_array, color='green', linestyle='None', marker='d') 
```

```python
plt.plot(random_array, color='green', linestyle=':', linewidth=3) ## colon (:) is for a dotted line
```

### line styles
 - default = solid
 - colon(:) = dotted
 - double dash(--)  = dashed
 - 'None' (with quotes as string) = no line only markers. Do specify marker property

## Highlighting the data points

#### using marker property
- 'd' - diamond shaped markers

#### marker size - size of markers (defualt = 6)

```python
plt.plot(random_array, color='green', marker='d')
```

# Scatter plots for bivariate plotting

```python
plt.scatter(range(0,20), random_array, color='green', marker='d') ## marker is optional
```
