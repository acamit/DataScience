```python
plt.plot(x, y)
plt.xlabel('Current')
plt.ylabel('Voltage')
plt.title('Ohms law')

plt.xlim([20, 80])
plt.ylim([200, 800])
```

```python
plt.plot(x, y, 'b+')  #blue + signs for points
plt.xlabel('Current')
plt.ylabel('Voltage')
plt.title('Ohms law')

plt.xlim([20, 80])
plt.ylim([200, 800])
```

```python
x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)
plt.plot(x, y, 'ro', x, y**2, 'b+', x, y**3, 'g^')  #blue + signs for points
plt.xlabel('Current')
plt.ylabel('Voltage')
plt.title('Ohms law')
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C,S = np.cos(X), np.sin(X)

plt.plot(X,C)
plt.plot(X,S)
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
```
