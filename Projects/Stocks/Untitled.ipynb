```python
metadata = metadata.dropna()
metadata.shape
```

```python
metadata['Year'] = metadata['to_date'].apply(lambda x:x.year)
```

```python
metadata['name'] = metadata['name'].apply(lambda x:re.sub('EOD Prices', '', x))
metadata.head()
```

```python
# metadata.loc[metadata.Year.isnull()]
```

```python
## get currently listed stocks
listedStocks = metadata.loc[metadata['Year'] == 2021]
listedStocks.reset_index(inplace=True)
print(listedStocks.shape)
listedStocks.head()
```

```python
# get data for first record manually this helps in initalizing the df as well. 
code = f'BSE/{listedStocks.loc[0, "code"]}'
name = listedStocks.loc[0, 'name']
bsecode = code.replace('BOM', '')
print(code, name, bsecode)
```

```python
dataset = quandl.get(code, start_date='2000-01-01', end_date='2021-01-21').sort_values(by='Date', ascending=True)
dataset['StockName']=name
dataset['BSECode'] = bsecode
```

```python
print(dataset.shape)
dataset.head()
```

```python
for i in range(1, len(listedStocks)):
    
    code = f'BSE/{listedStocks.loc[i, "code"]}'
    bsecode = code[7:]
    temp = quandl.get(code, start_date='2000-01-01', end_date='2021-03-16').sort_values(by='Date', ascending=True)
    temp['StockName']=listedStocks.loc[i, 'name']
    temp['BSECode'] = bsecode
    print(code, listedStocks.loc[i, 'name'], temp.shape)
    dataset = dataset.append(temp)
    print(dataset.shape)
    if i%10==0:
        dataset.to_csv('stocks.csv')
```
